```python
import matplotlib.pyplot as plt
import csv
import numpy as np

scale = np.arange(100, 3001, 100)
binary_small_0 = open("binary_104.csv", "r")
reader1 = csv.reader(binary_small_0)
all_lines = []
naive = []
kmp = []
optkmp = []
rk = []
for row in reader1:
    all_lines.append(row)
naive = list(map(int, all_lines[0][:-1]))
kmp = list(map(int, all_lines[1][:-1]))
optkmp = list(map(int, all_lines[2][:-1]))
rk = list(map(int, all_lines[3][:-1]))
plt.plot(scale, naive, label = 'BruteForce')
plt.plot(scale, kmp, label = 'KMP')
plt.plot(scale, optkmp, label = 'KMP + brs')
plt.plot(scale, rk, label = 'RK')
plt.grid(True)
plt.xlabel(' ')
plt.ylabel(', ')
plt.title('  10000    ')
plt.legend(loc = 'best')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
    
```
