```python
import matplotlib.pyplot as plt
import csv
import numpy as np

scale = np.arange(10, 1011, 50)
complete = open("complete.csv", "r")
reader = csv.reader(complete)
all_lines = []
for row in reader:
    all_lines.append(row)
dijkstra = list(map(int, all_lines[0][:-1]))
floydwarshall = list(map(int, all_lines[1][:-1]))
bellmanford = list(map(int, all_lines[2][:-1]))
obellmanford = list(map(int, all_lines[3][:-1]))
plt.plot(scale, dijkstra, label = 'Dijkstra')
plt.plot(scale, floydwarshall, label = 'Floyd-Warshall')
plt.plot(scale, bellmanford, label = 'Bellman-Ford')
plt.plot(scale, obellmanford, label = 'Optimized Bellman-Ford')
plt.grid(True)
plt.xlabel('   ')
plt.ylabel(', ')
plt.title(' ')
plt.legend(loc = 'best')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
scale = np.arange(10, 1011, 50)
all_lines = []
complete = open("complete.csv", "r")
reader = csv.reader(complete)
for row in reader:
    all_lines.append(row)
not_complete = open("not_complete.csv", "r")
reader = csv.reader(not_complete)
for row in reader:
    all_lines.append(row)
sparse = open("sparse.csv", "r")
reader = csv.reader(sparse)
for row in reader:
    all_lines.append(row)
result = [0] * 21
for i in range(21):
    s = 0
    for j in range(12):
        s += int(all_lines[j][i])
    result[i] = int(s / 12)
plt.plot(scale, result, label = 'average_alghoritms')
plt.grid(True)
plt.xlabel('   ')
plt.ylabel(', ')
plt.title('     ')
plt.legend(loc = 'best')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```
